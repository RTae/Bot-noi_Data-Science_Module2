```
def extractDec(html_page):
  description = html_page.select_one('div.entry-content-detail p span').text
  return description
```

```
def extract_warning(html_page):
    # find all p tags
    p_tags = html_page.find_all('p')
    # identify starting and ending p tags which correspond to warning and dosage paragraph
    for p in p_tags:
        print(p)
        if '' in p.text or ' Cylcysteine' in p.text:
            start = p
        if '' in p.text or '' in p.text or '' in p.text:
            end = p
            break
    # find all siblings below warnings
    sibs_down = start.find_next_siblings()
    # find all siblings above dosage
    sibs_up = end.find_previous_siblings()
    # find intersection
    warnings = [x for x in sibs_down if x in sibs_up]
    warnings_list = []
    if len(warnings) == 1:
      warnings_list = [ warning.text for warning in  warnings[0].select('li')]
    elif len(warnings) > 1 :
      for i in warnings:
        if len(i.find_all('li')) != 0:
          for j in i.select('li'):
            warnings_list.append(j.text)
    return warnings_list
```

```
def getDetail(url):
  response = requests.get(url)
  rawHTML = bs4.BeautifulSoup(response.content,'lxml')

  dec = extractDec(rawHTML)
  warming = extract_warning(rawHTML)
  return warming
```
